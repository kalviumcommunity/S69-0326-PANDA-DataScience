# 02 — ML Pipeline (Phase 2)

This notebook runs the complete ML pipeline built during Phase 2:

1. **Feature Engineering** — temporal, lag/rolling, business metrics, encoding
2. **Demand Forecasting** — Linear Regression, XGBoost, LightGBM
3. **Stockout Classification** — binary XGBoost
4. **Overstock Classification** — binary XGBoost
5. **Product Speed Classification** — multiclass XGBoost
6. **Product Clustering** — K-Means (k=3)
7. **Final Summary** — all model metrics side by side

```python
import sys, os

# Ensure project root is on path
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

print(f"Project root: {PROJECT_ROOT}")
```

---
## 1. Load Data & Feature Engineering

```python
from src.feature_engineering import build_features
from src.utils import get_data_path

raw = pd.read_csv(get_data_path())
print(f"Raw data shape: {raw.shape}")

df = build_features(raw)
print(f"\nEngineered data shape: {df.shape}")
print(f"Total features: {df.shape[1]}")
df.head()
```

---
## 2. Demand Forecasting

```python
from src.forecasting_model import train_forecast_model

forecast_results = train_forecast_model(df)
forecast_results
```

---
## 3. Stockout Classification

```python
from src.classification_model import train_stockout_classifier

stockout_metrics = train_stockout_classifier(df)
stockout_metrics
```

---
## 4. Overstock Classification

```python
from src.classification_model import train_overstock_classifier

overstock_metrics = train_overstock_classifier(df)
overstock_metrics
```

---
## 5. Product Speed Classification

```python
from src.classification_model import train_product_speed_classifier

speed_metrics = train_product_speed_classifier(df)
speed_metrics
```

---
## 6. Product Clustering

```python
from src.clustering_model import cluster_products

cluster_summary = cluster_products(df)
cluster_summary
```

---
## 7. Final Summary — All Model Metrics

```python
print("\n" + "=" * 70)
print("  FINAL MODEL PERFORMANCE SUMMARY")
print("=" * 70)

# ---- Forecasting ----
print("\n--- Demand Forecasting ---")
fc_df = pd.DataFrame(forecast_results).T
fc_df.index.name = "Model"
display(fc_df)

# ---- Classification ----
print("\n--- Classification Models ---")
clf_summary = pd.DataFrame({
    "Stockout": stockout_metrics,
    "Overstock": overstock_metrics,
}).T
display(clf_summary)

# ---- Product Speed ----
print("\n--- Product Speed (Multiclass) ---")
print(f"Macro F1: {speed_metrics['macro_f1']}")

# ---- Clustering ----
print("\n--- Product Clusters ---")
display(cluster_summary.groupby('cluster_label').size().rename('count'))

print("\n✅ Pipeline complete — all models saved in outputs/models/")
print("✅ All plots saved in outputs/plots/")
```
